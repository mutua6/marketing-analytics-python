```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (10, 6)

output_path = r'C:\Users\User\OneDrive\Documents\marketing-analytics-python\outputs'
os.makedirs(output_path, exist_ok=True)

# Load both datasets
df_ads = pd.read_csv(r'C:\Users\User\OneDrive\Documents\meta_ads_raw.csv')
df_crm = pd.read_csv(r'C:\Users\User\OneDrive\Documents\crm_customers.csv')

# Fix date columns
df_ads['date'] = pd.to_datetime(df_ads['date'], dayfirst=True)
df_crm['purchase_date'] = pd.to_datetime(df_crm['purchase_date'], dayfirst=True)

print("Ads shape:", df_ads.shape)
print("CRM shape:", df_crm.shape)
df_crm.head()
```

```python
# Each customer touched multiple campaigns before buying
# We simulate their journey based on campaign dates and purchase date

journeys = []

for _, customer in df_crm.iterrows():
    # Find all campaign touchpoints before purchase date
    touchpoints = df_ads[
        df_ads['date'] <= customer['purchase_date']
    ][['campaign_name', 'date']].copy()
    
    touchpoints['customer_id'] = customer['customer_id']
    touchpoints['purchase_value'] = customer['purchase_value']
    touchpoints['purchase_date'] = customer['purchase_date']
    touchpoints = touchpoints.sort_values('date')
    
    journeys.append(touchpoints)

df_journey = pd.concat(journeys, ignore_index=True)

print("Journey shape:", df_journey.shape)
print("\nSample customer journey:")
df_journey[df_journey['customer_id'] == 'C001']
```
